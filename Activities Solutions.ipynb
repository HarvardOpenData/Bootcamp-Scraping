```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.hbs.edu/faculty/Pages/browse.aspx"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
```

```python
# Scrape the links to HBS faculty profile pages by finding where they're located on the HTML pages and then using "find_all" and "get"!
# Hint: You may want to use multiple for loops while trying to get the links

# First, call the "find_all" function to get all "faculty-item" divs
all_faculty = soup.find_all("div", attrs = {"class" : "faculty-item"})

#Next, we can find the links contained in each faculty-item using a couple of for loops:
for faculty in all_faculty:
    links = faculty.find_all('a')
    for link in links:
        print(link.get('href'))
```

```python
# Extra imaginary points if you can print the links with the hbs.edu appended at the beginning

# Same process as above, but we can also just append"hbs.edu" to the start of the hrefs totally not necessary to know the "replace" Python function off the top of your head though!
all_faculty = soup.find_all("div", attrs = {"class" : "faculty-item"})
for faculty in all_faculty:
    tests = faculty.find_all('a')
    for test in tests:
        print(("hbs.edu" + test.get('href')).replace("\n", "").replace(" ", ""))
```

```python
# Scrape the Open Secrets 2020 election contribution webpage and find the percentage of the top 10 election contributions that went to Democrats vs. Republicans

url = "https://www.opensecrets.org/overview/topindivs.php?cycle=2020&view=fc"

# We can disguise our request
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
# gcontext = ssl.SSLContext()
html = urlopen(req).read()

soup = BeautifulSoup(html, 'lxml')

# Notice that we disguised our request when accessing the Open Secrets website due to increased security

election2020 = soup.find_all("td", attrs = {"class" : "number pct"})
```

```python
# Note: Re-assigning these variables will override their previous definitions when scraping HBS Faculty data!
```

```python
# We start by defining a couple of variables for the total amounts of money donated to Democrats and Republicans
dem_sum = 0
rep_sum = 0

# Next, we can use a combination of the find_all, get_text, and replace functions to get only the numbers we want from the table, which are stored in the 4th and 5th positions of each row
for tr in (soup.find_all('tr')[1:11]):
    cleantextdem = (tr.find_all('td', {'class': 'number'})[3]).get_text(strip = True)
    dem_sum = dem_sum + float(cleantextdem.replace('$', '').replace(',', ''))
    cleantextrep = (tr.find_all('td', {'class': 'number'})[4]).get_text(strip = True)
    rep_sum = rep_sum + float(cleantextrep.replace('$', '').replace(',', ''))

# Lastly, we just need to do a bit of match to get the percentages of contributions that went to each party.
print("Democrats: " + str(dem_sum/(dem_sum+rep_sum)*100) +
" , Republicans: " + str(rep_sum/(dem_sum+rep_sum)*100))
```

```python
# We can start by defining a function containing our scraper (this is the exact same code as before, just in a Python function):

def scraper(url):
    # We can disguise our request
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    # gcontext = ssl.SSLContext()
    html = urlopen(req).read()

    soup = BeautifulSoup(html, 'lxml')

    dem_sum = 0
    rep_sum = 0

    for tr in (soup.find_all('tr')[1:11]):
        cleantextdem = (tr.find_all('td', {'class': 'number'})[3]).get_text(strip = True)
        dem_sum = dem_sum + float(cleantextdem.replace('$', '').replace(',', ''))
        cleantextrep = (tr.find_all('td', {'class': 'number'})[4]).get_text(strip = True)
        rep_sum = rep_sum + float(cleantextrep.replace('$', '').replace(',', ''))

    return("Democrats: " + str(dem_sum/(dem_sum+rep_sum)*100) +
    " , Republicans: " + str(rep_sum/(dem_sum+rep_sum)*100))
```

```python
# Next, we can take into account the fact that only the year changes in each election cycle's OpenSecrets URL and create a URL generator function:

def url_generator(year):
    return("https://www.opensecrets.org/elections-overview/biggest-donors?cycle=" + str(year) + "&view=fc")
```

```python
# Lastly, we can take the years in question and find the percentages for each election:

years = [2014, 2016, 2018, 2020]
for year in years:
    print(scraper(url_generator(year)))
```
